```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.signal
```

# Dados da Tabela do Caio

|$V_a$|$I_a$|$n$|$V_{taco}$|
|--|
|230|1.0|1466|88|
|167|0.8|1075|64.4|
|103|0.75|670|40.2|
|42|0.65|270|16.1|

```python
Va    = np.array([230, 167, 103, 42])
Ia    = np.array([1.0, 0.8, 0.75, 0.65])
n     = np.array([1466, 1075, 670, 270])
Vtaco = np.array([88, 64.4, 40.2, 16.1])
```

# Cálculo do Ra e Xa

```python
Vcc = 2.1
Icc = 1.99
Vca = 16.2
Ica = 2.01

Ra = Vcc / Icc
Xa = np.sqrt((Vca/Ica)**2 - Ra**2)
print('Ra =', Ra, 'Xa =', Xa)
```

# Cálculo do $K_\phi$

```python
w = n * 2*np.pi/60
Kphi = (Va - Ra*Ia) / w
print('Kphi =', Kphi)
print('Kphi médio =', np.mean(Kphi))
```

# Cálculo do B

```python
Tnl = Kphi*Ia
B = Tnl/w
print('B =', B)
print('B médio =', np.mean(B))
```

# Cálculo do J

```python
data = pd.read_csv('values.csv')
data.y *= 10 # pra compensar a atenuação da ponta de prova
plt.plot(data.x, data.y)
plt.title('Curva de Desaceleração')
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão no Tacogerador (V)')
plt.show()
```

```python
# Cortando as partes estranhas
data = data[data.x < 15] # Corta a parte = 0
data = data.ix[100:, :] # Corta a parte plana no topo
fator = np.mean(n/ Vtaco) # Fator de conversão do tacogerador
data.w = data.y * fator * 2*np.pi/60
plt.plot(data.x, data.w)
plt.title('Curva de Desaceleração')
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade Angular do Rotor (rad/s)')
plt.show()
```

```python
# Filtra o sinal
fs = 1/(np.mean(np.diff(data.x))) # frequencia de amostragem
fnyq = fs * .5 # frequencia de nyquist
fc = 10 # frequencia de corte (Hz)
num, denom = scipy.signal.butter(N=4, Wn=fc/fnyq) # passa-baixa butterworth de 4ª ordem
data['w_filtered'] = scipy.signal.lfilter(num, denom, data.w)
normal = plt.plot(data.x, data.w, label = 'Sinal não filtrado')
filtrado = plt.plot(data.x, data.w_filtered, color='red', label='Sinal filtrado')
plt.legend()
plt.title('Curva de Desaceleração')
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade Angular do Rotor (rad/s)')
plt.show()
```

```python
# Diferencia o sinal filtrado
dw = np.diff(data.w_filtered) / np.diff(data.x)
dw = dw[np.abs(dw) < 20] # corta o transiente
dw_medio = np.mean(dw)
print('dw/dt =', dw_medio)
```

```python
J = -B*w / dw_medio
print('J =', J)
print('J médio =', np.mean(J))
```
